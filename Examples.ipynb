```python
graph = nx.read_graphml('/var/essdata/IDU/other/CityGeoToolsStandalone/scr/CityGeoTools/examples/spb_exmpl_MobilityGraph.graphml', 
                        node_type=int)
graph = nx.relabel_nodes(graph, (dict(zip(graph.nodes,
                                 [int(x) for x in graph.nodes]))))
for i, n in graph.nodes(data=True):
    n['geometry'] = shapely.geometry.Point(n['x'], n['y'])
```
